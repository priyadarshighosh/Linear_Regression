# Main our Own Linear Regression Class

```python
class MeraLR:

   def __init__(self):        #this is the default constructor that assigns None value to coef and intercept
      self.coef_ = None
      self.intercept_ = None

   def fit(self,X_train,y_train):
      X_train = np.insert(X_train,0,1,axis=1)     # where are we filling the dataset that is X_train , position that we are filling that is  0 , what are we filling that is 1 as we are taking the value of Beta0 as 1 , and which axis are we filling if its rows then axis=0 if its columns then axis=1 , we are making a new column so axis will be 1

      # calculate the coefficents
      betas = np.linalg.inv(np.dot(X_train.T,X_train)).dot(X_train.T).dot(y_train)   # this calculates the beta
      self.intercept_ = betas[0]
      self.coef_ = betas[1:]
      print(betas)


   def predict(self,X_test):
       y_pred = np.dot(X_test,self.coef_) + self.intercept_     # We are basically multiplying the 2 matrices X_tests that has the inputs and  self.coefs that has the weights of the inputs and adding BETA0 to the mix in the form of intercept as
       return y_pred

```
