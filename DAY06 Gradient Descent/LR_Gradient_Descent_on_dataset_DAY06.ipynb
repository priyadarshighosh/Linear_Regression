#  Gradient Descent REGRESSION

```python
class GDRegressor:

  def __init__(self , learning_rate , epochs):     # we are initializing the values as random weird values
      self.m = 100
      self.b = -120

      self.lr = learning_rate
      self.epochs = epochs

  def fit(self, X , y ):

    for i in range(self.epochs):          # calculating the gradient descent  changing the  b and  m

      loss_slope_b = -2 * np.sum(y - self.m*X.ravel() - self.b )
      self.b = self.b  - (self.lr * loss_slope_b)

      loss_slope_m = -2 * np.sum((y - self.m*X.ravel() - self.b ) * X.ravel())
      self.m = self.m - (self.lr * loss_slope_m)

    print(self.m  , self.b)

  def predict(self , X):
     return self.m * X + self.b




```

Making an object of the GD Regressor Class


```python
gd = GDRegressor(0.001 , 150)
```

```python
gd.fit (X_train , y_train)
```
